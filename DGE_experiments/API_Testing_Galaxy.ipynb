## 2nd API for testing: Data collections

#If you want, you can list existing data collections within a specific history

#First, list all the histories

```python
histories = gi.histories.get_histories()
print("Histories:", histories)
```

#List all the collections in the selected history, in our case, let's do a786ace1e9db0920

```python
history_id="a786ace1e9db0920"
history_contents = gi.histories.show_history(history_id, contents=True)
collections = [item for item in history_contents if item['history_content_type'] == 'dataset_collection']
print("Dataset Collections in History:", collections)
```

#Now let's try to create a new dataset collection with individual datasets.

#to identify the dataset ID to include in the collection:

```python
history_contents = gi.histories.show_history(history_id, contents=True)
datasets = [item for item in history_contents if item['history_content_type'] == 'dataset']
print("Datasets in History:", datasets)
```

#Now, you want to create the dataset collection, and to do so, we want to create a collection like so: 

```python
dataset_ids = ['f9cad7b01a4721351f9155753efdfa07', 'f9cad7b01a4721356bcfdb55db735da2']
element_identifiers = [{'id': dataset_id, 'name': f"Dataset_{index + 1}", 'src': 'hda'} for index, dataset_id in enumerate(dataset_ids)]
collection_name = "Test Collection"
new_collection = gi.histories.create_dataset_collection(
    history_id,
    {
        'name': collection_name,  # Add the 'name' key here
        'collection_type': 'list',
        'element_identifiers': element_identifiers
    }
)
print("Created Collection:", new_collection)
```

#Now, we can retrieve the details of a specific dataset collection

```python
collection_id = 'd9f1c80e95f60ba3'
collection_details = gi.histories.show_dataset_collection(history_id, collection_id)
print("Collection Details:", collection_details)
```
